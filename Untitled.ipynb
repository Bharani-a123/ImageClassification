```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from collections import Counter

# ============================
# 1. ADVANCED AUGMENTATION
# ============================
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    brightness_range=[0.7, 1.3],  # helps reduce similarity confusion
    channel_shift_range=30.0,      # enhances feature differences
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest',
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    "modified-dataset/train",
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical",
    subset="training"
)

val_genertator = train_datagen.flow_from_directory(
    "modified-dataset/train",
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical",
    subset="validation"
)

# ============================
# 2. TRANSFER LEARNING MODEL
# ============================
base_model = EfficientNetB0(weights="imagenet", include_top=False, input_shape=(224,224,3))
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x)
preds = Dense(train_generator.num_classes, activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=preds)

for layer in base_model.layers:
    layer.trainable = False  # freeze base layers first

model.compile(optimizer=Adam(learning_rate=0.0005), loss="categorical_crossentropy", metrics=["accuracy"])

# ============================
# 3. TRAIN MODEL (for noise detection)
# ============================
history = model.fit(
    train_generator,
    epochs=5,  # small training first
    validation_data=val_generator
)

# ============================
# 4. NOISE / MISLABEL DETECTION
# ============================
# Get predictions for training set
train_generator.reset()
preds = model.predict(train_generator, verbose=1)
pred_classes = np.argmax(preds, axis=1)
true_classes = train_generator.classes
class_labels = list(train_generator.class_indices.keys())

# Find mismatched predictions (possible mislabels)
mismatches = np.where(pred_classes != true_classes)[0]

print(f"Total potential noisy/mislabeled images: {len(mismatches)}")

# ============================
# 5. VISUALIZE SUSPECT IMAGES
# ============================
def show_mislabeled_samples(generator, mismatches, n=10):
    for i, idx in enumerate(mismatches[:n]):
        img, label = generator[idx]
        plt.imshow(img[0])
        plt.title(f"True: {class_labels[label[0].argmax()]} | Pred: {class_labels[pred_classes[idx]]}")
        plt.axis("off")
        plt.show()

show_mislabeled_samples(train_generator, mismatches, n=5)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam

# ✅ Normalize images properly [Fix for Error 2]
train_datagen = ImageDataGenerator(
    rescale=1./255,        # normalization
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)  # ✅ normalization here too

train_generator = train_datagen.flow_from_directory(
    'modified-dataset/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    'modified-dataset/val',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

# ✅ Save correct class indices [Fix for Error 5]
class_indices = train_generator.class_indices
labels = {v: k for k, v in class_indices.items()}
print("Class mapping:", labels)

# Build model (transfer learning with VGG16)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224,224,3))
x = Flatten()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
preds = Dense(len(class_indices), activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=preds)

for layer in base_model.layers:
    layer.trainable = False

model.compile(optimizer=Adam(learning_rate=1e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)

# ✅ Correct prediction decoding
import numpy as np
from tensorflow.keras.preprocessing import image

def predict_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img) / 255.0  # normalization
    img_array = np.expand_dims(img_array, axis=0)

    pred = model.predict(img_array)
    class_id = np.argmax(pred)
    return labels[class_id]  # ✅ Correct mapping

print("Prediction:", predict_image("test_img.jpg"))
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

# ✅ Function to predict on single image
def predict_image(img_path, model, labels):
    img = image.load_img(img_path, target_size=(224, 224))  # Resize to model input
    img_array = image.img_to_array(img) / 255.0             # Normalize
    img_array = np.expand_dims(img_array, axis=0)           # Add batch dimension
    
    pred = model.predict(img_array)
    class_id = np.argmax(pred)   # Highest probability
    confidence = np.max(pred)    # Confidence score
    
    return labels[class_id], confidence

# Example usage
labels = list(train_generator.class_indices.keys())  # ✅ Get class labels from training
img_path = r"C:\Users\BHARANIDHARAN.S\Dailyproject/IMG/modified-dataset/test/Battery/battery_174.jpg"# <-- Replace with your image path
predicted_class, confidence = predict_image(img_path, model, labels)

print(f"Prediction: {predicted_class} ({confidence*100:.2f}% confidence)")
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ✅ Load trained model
model = load_model("my_model.keras")

# ✅ Generators with same preprocessing
datagen = ImageDataGenerator(rescale=1./255)

train_generator = datagen.flow_from_directory(
    'modified-dataset/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

val_generator = datagen.flow_from_directory(
    'modified-dataset/val',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# ✅ Evaluate on train and val
train_loss, train_acc = model.evaluate(train_generator, verbose=1)
val_loss, val_acc = model.evaluate(val_generator, verbose=1)

print("\n📊 Model Performance Check:")
print(f"✅ Train Accuracy: {train_acc*100:.2f}% | Train Loss: {train_loss:.4f}")
print(f"✅ Val Accuracy:   {val_acc*100:.2f}% | Val Loss:   {val_loss:.4f}")

# ✅ Overfitting check
if train_acc - val_acc > 0.10:   # >10% gap means overfitting
    print("⚠️ Model may be OVERFITTED!")
else:
    print("✅ Model generalizes well.")
```
