```python
import zipfile
import os

# Path to your ZIP file
zip_path = "imageClassificationdEwast data.zip"
extract_path = "unzipped_dataset"  # This folder will be created

# Unzipping
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("✅ Dataset unzipped to:", extract_path)
```

```python
from PIL import Image
import os

def get_image_counts_by_class(extract_path):
    image_counts = {}
    corrupted_images = []

    for split in ['train', 'val', 'test']:
        split_path = os.path.join(extract_path, split)
        if not os.path.exists(split_path):
            continue  # Skip if the folder doesn't exist

        for root, dirs, files in os.walk(split_path):
            for file in files:
                if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                    file_path = os.path.join(root, file)
                    try:
                        img = Image.open(file_path)
                        img.verify()  # Verify image is not corrupted

                        class_name = os.path.basename(os.path.dirname(file_path))
                        image_counts[class_name] = image_counts.get(class_name, 0) + 1
                    except:
                        corrupted_images.append(file_path)

    return image_counts, corrupted_images

# Replace with your actual path
extract_path = '/mnt/data/imageClassificationdEwast data'
image_counts, corrupted_images = get_image_counts_by_class(extract_path)
```

```python
import matplotlib.pyplot as plt

# image_counts is already returned from check_dataset()
# But if you are running fresh, rerun check_dataset() first.

def plot_class_distribution(image_counts):
    classes = list(image_counts.keys())
    counts = list(image_counts.values())

    plt.figure(figsize=(10, 6))
    plt.bar(classes, counts, color='skyblue', edgecolor='black')
    plt.xlabel('Class Name')
    plt.ylabel('Number of Images')
    plt.title('Image Count per Class')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

plot_class_distribution(image_counts)
```
